```python
#Qual Ã© a probabilidade de encontrar um cliente na RegiÃ£o Sul e na Classe Residencial?


sul_residencial = df.loc["Sul", "Residencial"]
total_geral = df.values.sum()
prob_sul_residencial = sul_residencial / total_geral
print(f"\nProbabilidade de encontrar um cliente na RegiÃ£o Sul e Residencial: {prob_sul_residencial}")
```

```python
#Qual Ã© a probabilidade de encontrar um cliente na RegiÃ£o Sudeste dado que seja da classe Comercial?


sudeste_comercial = df.loc["Sudeste", "Comercial"]
total_comercial = df["Comercial"].sum()
prob_sudeste_dado_comercial = sudeste_comercial / total_comercial
print(f"\nProbabilidade de cliente na RegiÃ£o Sudeste dado que seja Comercial: {prob_sudeste_dado_comercial:.2f}")
```

```python
#Qual a funÃ§Ã£o de custo (x)?

import sympy as sp

#Variavel X
x = sp.symbols('x')

#A) FunÃ§Ã£o de custo C(x)
C_x = 25000 + 5000*x**2 - 20000*x
print("FunÃ§Ã£o de Custo C(x):")
print(C_x)
```

```python
#Qual a funÃ§Ã£o de receita/faturamento R(x)?


R_x = 25000*x + 15000
print("\nFunÃ§Ã£o de Receita R(x):")
print(R_x)
```

```python
#Qual a quantidade aproximada de instalaÃ§Ãµes para que a empresa atinja o ponto de equilÃ­brio (Receitas = Custos)?

#VariÃ¡vel x
x = sp.symbols('x')

# Coeficientes da equaÃ§Ã£o 5.000x^2 - 20.000x - 10.000 = 0
a = 5000
b = -20000
c = -10000

#EquaÃ§Ã£o: a*x^2 + b*x + c = 0
equacao = sp.Eq(a*x**2 + b*x + c, 0)

#Calculo da equaÃ§Ã£o
solucoes = sp.solve(equacao, x)

print("SoluÃ§Ãµes para o ponto de equilÃ­brio (R(x) = C(x)): ")
for sol in solucoes:
    print(f"x = {sol.evalf():.2f}")

print("R: A quantidade para o ponto de Ã©quilibrio Ã© de 5 unidades")
```

```python
#Qual a quantidade de instalaÃ§Ãµes que minimizam o custo C(x)?


C_prime_x = sp.diff(C_x, x)
solucoes_minimo = sp.solve(sp.Eq(C_prime_x, 0), x)
print("\nSoluÃ§Ãµes para o ponto que minimiza o custo C(x):")
print(solucoes_minimo)
print("R: A quantidade que minimiza o custo Ã© de 2 unidades")
```
